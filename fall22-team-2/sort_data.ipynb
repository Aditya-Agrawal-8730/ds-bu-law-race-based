```python
# import list of libraries

import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('inventor_data with patent number and claims number.csv')
df[df['inventor_country_code'].isin(['KR', 'CN'])].to_csv('inventors_filtered_by_cn_kr.csv', index=False)
```

```python
cn_kr_inventors = pd.read_csv('inventors_filtered_by_cn_kr.csv')
cn_kr_inventors.loc[cn_kr_inventors['gender'].isin(['male', 'female', 'andy']) == False, 'gender'] = np.NaN
cn_kr_inventors.loc[cn_kr_inventors['gender'] == 'andy', 'gender'] = 'androgynous'
cn_kr_inventors.loc[cn_kr_inventors['gender'].isin(['male', 'female', 'androgynous']) == False]
```

```python
language_index = pd.read_csv('wgnd_2_0_code-langcode.csv')
reg = re.compile(r'[a-zA-Z]')

data_gender_determined = pd.read_csv('wgnd_2_0_name-gender-langcode.csv')
zhko = data_gender_determined.loc[data_gender_determined['langcode'].isin(['zh', 'ko'])]
zhko = zhko.loc[zhko['name'].str.match(reg) == False]
zhko.to_csv('zhko.csv', index=False)

data_with_weight = pd.read_csv('wgnd_2_0_name-gender-code.csv')
zhko_country_list = language_index.loc[language_index['langcode'].isin(['zh', 'ko'])]['code'].to_list()
data_with_weight = data_with_weight.loc[data_with_weight['code'].isin(zhko_country_list)]
zhko_with_weight = data_with_weight.loc[data_with_weight['name'].str.match(reg) == False]
zhko_with_weight.to_csv('zhko_with_weight.csv', index=False)
```
